```python
import os
from ultralytics import YOLO
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import UserMessage
from azure.core.credentials import AzureKeyCredential

# --- 1. Load YOLO Model ---
yolo_model = YOLO("/content/best.pt")  # model jerawat kamu

# --- 2. Deteksi jerawat dari gambar ---
image_path = "/content/WHITEHEADS DAN PUSTULE.jpg"
results = yolo_model(image_path)[0]

# Ambil nama kelas jerawat yang terdeteksi
class_ids = results.boxes.cls.cpu().numpy()
class_names = results.names
detected_acne = [class_names[int(i)] for i in class_ids]

# Buat ringkasan hasil deteksi
summary = {}
for acne in detected_acne:
    summary[acne] = summary.get(acne, 0) + 1

summary_text = ", ".join([f"{count} {name}" for name, count in summary.items()])

print("Hasil deteksi:", summary_text)

# --- 3. Panggil LLM GitHub untuk rekomendasi skincare ---
client = ChatCompletionsClient(
    endpoint="https://models.github.ai/inference",
    credential=AzureKeyCredential(os.environ["GITHUB_TOKEN"]),  # pastikan token GitHub diset
)

prompt = f"""
Saya mendeteksi kondisi kulit dengan hasil: {summary_text}.
Tolong berikan penjelasan singkat mengenai kondisi ini
dan rekomendasi skincare dasar yang sesuai (misalnya cleanser, serum, treatment).
Gunakan bahasa yang mudah dipahami.
"""

response = client.complete(
    messages=[UserMessage(prompt)],
    model="gpt-4o-mini",  # bisa diganti "meta-llama/Llama-3-8B-Instruct" jika tersedia
    max_tokens=300,
)

print("\n--- Rekomendasi Skincare ---")
print(response.choices[0].message.content)
```

```python
import os
from ultralytics import YOLO
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import UserMessage
from azure.core.credentials import AzureKeyCredential
import cv2
import matplotlib.pyplot as plt

# ===============================
# 1. Load YOLO Model
# ===============================
yolo_model = YOLO("/content/best.pt")  # ganti path modelmu

# Gambar input
image_path = "/content/WHITEHEADS DAN PUSTULE.jpg"
results = yolo_model(image_path)[0]

# ===============================
# 2. Hitung jumlah jerawat
# ===============================
class_ids = results.boxes.cls.cpu().numpy()
class_names = results.names
detected_acne = [class_names[int(i)] for i in class_ids]

summary = {}
for acne in detected_acne:
    summary[acne] = summary.get(acne, 0) + 1

summary_text = ", ".join([f"{count} {name}" for name, count in summary.items()])
print("Hasil deteksi:", summary_text)

# ===============================
# 3. Tampilkan gambar hasil deteksi
# ===============================
annotated_img = results.plot()

plt.figure(figsize=(8,8))
plt.imshow(cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Hasil Deteksi Jerawat")
plt.show()

# ===============================
# 4. Hubungkan ke LLM (Azure GitHub Models)
# ===============================
client = ChatCompletionsClient(
    endpoint="https://models.github.ai/inference",
    credential=AzureKeyCredential(os.environ["GITHUB_TOKEN"]),  # pastikan token GitHub sudah diset
)

prompt = f"""
Saya mendeteksi kondisi kulit dengan hasil: {summary_text}.
Tolong berikan penjelasan singkat mengenai kondisi ini
dan rekomendasi skincare dasar yang sesuai (misalnya cleanser, serum, treatment).
Gunakan bahasa yang mudah dipahami.
"""

response = client.complete(
    messages=[UserMessage(prompt)],
    model="gpt-4o-mini",  # bisa diganti model lain jika tersedia
    max_tokens=300,
)

print("\n--- Rekomendasi Skincare ---")
print(response.choices[0].message.content)
```
